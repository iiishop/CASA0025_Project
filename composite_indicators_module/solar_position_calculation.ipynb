```python
import pandas as pd
from pvlib.solarposition import get_solarposition
```

```python
# Study area: London
lat = 51.5072
lon = -0.1276
timezone = "Europe/London"

# Analysis period and sampling interval
times = pd.date_range(
    start="2025-04-19 06:00",
    end="2026-09-19 20:00",
    freq="1h",
    tz=timezone
)

# Calculate solar position
solar = get_solarposition(
    time=times,
    latitude=lat,
    longitude=lon
)

# Keep only daytime observations
daytime = solar[solar["apparent_elevation"] > 0]

# Mean solar angles across the whole period
mean_azimuth = daytime["azimuth"].mean()
mean_zenith = daytime["apparent_zenith"].mean()

print("Mean azimuth:", mean_azimuth)
print("Mean zenith:", mean_zenith)
```
